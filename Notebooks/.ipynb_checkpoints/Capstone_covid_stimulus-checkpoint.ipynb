```python
snp_500.columns=['Date', 'S&P 500 (TR)' , 'S&P 500 (Net TR)' , 'S&P 500']
snp_500
```

```python
snp_500.Date = pd.to_datetime(snp_500.Date)
snp_500.Date = snp_500.Date.dt.strftime('%m/%d/%Y')
```

```python
snp_500
```

```python
merge_snp_500_UMP = pd.merge(left=snp_500,right=unemp_rate,left_on='Date',right_on='Date',how='outer')
merge_snp_500_UMP
```

```python
merge_snp_500_UMP_real_Hrly_Earn = pd.merge(left=merge_snp_500_UMP,right=real_hrly_earn,left_on='Date',right_on='Date',how='outer')
merge_snp_500_UMP_real_Hrly_Earn
```

```python
merge_snp_500_UMP_real_Hrly_Earn_real_gdp = pd.merge(left=merge_snp_500_UMP_real_Hrly_Earn,right=real_gdp,left_on='Date',right_on='Date',how='outer')
merge_snp_500_UMP_real_Hrly_Earn_real_gdp
```

```python
merge_snp_500_UMP_real_Hrly_Earn_real_gdp_DTY_curve_rates = pd.merge(left=merge_snp_500_UMP_real_Hrly_Earn_real_gdp,right=DTY_curve_rates,left_on='Date',right_on='Date',how='outer')
merge_snp_500_UMP_real_Hrly_Earn_real_gdp_DTY_curve_rates
```

```python
merged_data = merge_snp_500_UMP_real_Hrly_Earn_real_gdp_DTY_curve_rates
```

```python
merged_data = pd.merge(left=merged_data,right=PPP_industry,left_on='Date',right_on='Date',how='outer')
merged_data
```
